### 1.1. Скачать данные

```python
df = pd.read_csv('kc_house_data.csv')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize = (8,6))
sns.histplot(df['price'])
plt.xlabel('Цена')
plt.ylabel('Кол-во домов')
plt.title('Стоимость недвижимости')
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize = (8,6))
plt.hist(df['sqft_living'], bins=15)
plt.xlabel('Жилая площадь')
plt.ylabel('Количество домов')
plt.title('Квадратура жилой площади')
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize = (10,6))
plt.hist(df['yr_built'], bins=15)
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.title('Даты строительства домов')
```

## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg') 
```

```python
# По графику видно, что площадь напрямую влияет на стоимость жилья. Чем выше площадь, тем больше стоимость, но и спрос меньше
```

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg') 
```

```python
# стоимость дома увеличивается в зависимости от его новизны
```

```python
sns.jointplot(x=df['price'], y=df['waterfront'], kind='reg') 
```

```python
# Зависимости стоимости дома от вида на набережную не наблюдается
```

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg') 
```

```python
# Этажность дома не сильно влияет на его стоимость. Есть и более дорогие дома в низкой этажности и менее дорогие в высокой этажности
```

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='reg') 
```

```python
# Дома в более хорошем состоянии имеют большую стоимость, но самые дорогие дома встречаются большей части среди домов в хорошем состоянии
```
